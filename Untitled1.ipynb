# <en><center>Libraries</center></en>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
```

# <en><center>Prepare Data</center></en>

## Import Data

```python
def wrangle(filepath):
    df = pd.read_csv(filepath)
    
    return df
```

```python
df = wrangle("California_Houses.csv")
```

## Data Exploration

```python
df.info()
```

```python
df.head()
```

```python
for col in df.columns:
    print(col)
```

```python

sns.displot(
    df["Median_House_Value"],
    bins = 30,
    color = "red",
    alpha = 0.5,
    aspect = 2
)

plt.axvline(
    np.mean(df["Median_House_Value"]),
    color = "blue",
    ls = "--"
)
plt.axvline(
    np.median(df["Median_House_Value"]),
    color = "green",
    ls = ":"
);
```

```python
df["Median_House_Value"].describe()
```

Top 10% Median House Values

```python
high_mask = df["Median_House_Value"] > 400_000
```

```python
df[high_mask]["Median_House_Value"].describe()
```

```python
sns.displot(
    df[high_mask]["Median_House_Value"],
    color = "orange",
    alpha = 0.5,
    aspect = 2
);
```

```python
fig = px.scatter_mapbox(
    df,
    lat = "Latitude",
    lon = "Longitude",
    width = 800,
    height = 800,
    color = "Median_House_Value",
    hover_data = ["Median_House_Value"]
) 
fig.update_layout(mapbox_style = "open-street-map")
fig.show()
```
